## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata_df = pd.read_csv(mouse_metadata_path)
study_results_df = pd.read_csv(study_results_path)

# Combine the data into a single dataset
pyma_df = pd.merge(mouse_metadata_df,study_results_df, on="Mouse ID", how="left")
# Display the data table for preview
pyma_df.head()
```

```python
# Checking the number of mice.
mouse_metadata_df['Mouse ID'].nunique
```

```python

drop_dup_id = study_results_df.loc [study_results_df.duplicated(subset=['Mouse ID','Timepoint',]),'Mouse ID'].unique()
clean_study_results_df = study_results_df[study_results_df['Mouse ID'].isin(drop_dup_id)==False] 
clean_mouse_metadata_df = mouse_metadata_df[mouse_metadata_df['Mouse ID'].isin(drop_dup_id)==False]

combined_data = pd.merge(clean_study_results_df, clean_mouse_metadata_df, on = "Mouse ID")

combined_data
```

```python


combined_data_sort = combined_data.sort_values(["Tumor Volume (mm3)"],ascending=True)

combined_data_sort
```

```python
study_results_df.sort_values("Mouse ID", inplace = True)
study_results_df.drop_duplicates(subset ="Mouse ID", keep = False, inplace = True) 

study_results_df
clean_df =  pd.merge(mouse_metadata_df,study_results_df, on="Mouse ID", how="left")
clean_df 
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
regimen_group = combined_data_sort.groupby(["Drug Regimen"])
regimen_group

regimen_mean = regimen_group["Tumor Volume (mm3)"].mean()
regimen_median = regimen_group["Tumor Volume (mm3)"].median()
regimen_variance = regimen_group["Tumor Volume (mm3)"].var()
regimen_std = regimen_group["Tumor Volume (mm3)"].std()
regimen_sem = regimen_group["Tumor Volume (mm3)"].sem()
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_stats = pd.DataFrame({"Mean": regimen_mean, "Median":regimen_median, "Variance":regimen_variance, "Standard Deviation": regimen_std, "SEM": regimen_sem})
summary_stats
# Using the aggregation method, produce the same summary statistics in a single line
```
